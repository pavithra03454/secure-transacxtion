```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading datasets as a pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard(1).csv')
```

```python
credit_card_data.isnull().sum()
```

```python
#Handling Null values
#Filling Null values
credit_card_data_2 = credit_card_data.fillna(value=0)
```

```python
credit_card_data_2.isnull().sum().sum()
```

```python
#distribution of legit transcations & fraud
credit_card_data_2['Class'].value_counts()
```

This data set is highly unbalanced

1.   0---normal transaction
2.   1---fradulent transaction



```python
#separating legit and fraudlent data
legit = credit_card_data_2[credit_card_data_2.Class == 0]
fraud = credit_card_data_2[credit_card_data_2.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare  the values for both transactions
credit_card_data_2.groupby('Class').mean()
```

Under sampling

*   build the sample dataset conntaining similar distribution of normal transactions and fradulent




Number of fraudlent transaction

```python
legit_sample = legit.sample(n = 492 )
```

concatinating two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting the data into features & targets

```python
x = new_dataset.drop(columns = 'Class' , axis = 1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

split the data into training data & testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y ,random_state=2)
```

```python
print(x.shape ,x_train.shape ,x_test.shape)
```

model training

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

model evaluation

accuarcy score

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction , y_train)
```

```python
print('accuracy on training data:',training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction , y_test)
```

```python
print('accuracy on test data:',testing_data_accuracy)
```
